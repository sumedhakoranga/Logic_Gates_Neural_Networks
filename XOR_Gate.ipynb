# Initialization

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import random
```

# Generating synthetic data

```python
p0 = [[0, 0, 0] for _ in range(random.randint(100, 150))]
p1 = [[0,1,1] for _ in range(random.randint(100, 150))]
p2 = [[1,0,1] for _ in range(random.randint(100, 150))]
p3 = [[1,1,0] for _ in range(random.randint(100, 150))]
X = np.array(p0+p1+p2+p3)
np.random.shuffle(X)
y = X[:, 2]
X = X[:, :2]
```

```python
X
```

```python
y
```

# Training the model

```python
model = keras.Sequential([
    keras.Input(shape=(2)),
    keras.layers.Dense(2,activation="sigmoid"),
    keras.layers.Dense(1,activation="sigmoid")]
)
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics='accuracy')
```

```python
model.fit(X,y,epochs=1000,batch_size=32)
```
